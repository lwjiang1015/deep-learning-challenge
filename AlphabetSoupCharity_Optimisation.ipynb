```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from scipy.stats.mstats import winsorize
```

```python
#  Import and read the charity_data.csv.

application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df =application_df.drop(columns = ["EIN", "NAME"])
application_df.head()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
type_value_counts = application_df["APPLICATION_TYPE"].value_counts()
type_value_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

application_types_to_replace = type_value_counts[type_value_counts<528].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
cls_value_counts = application_df["CLASSIFICATION"].value_counts()
cls_value_counts
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
filtered_cls_value_counts_to_replace = cls_value_counts[cls_value_counts<1883]
classifications_to_replace = filtered_cls_value_counts_to_replace.index.tolist()
print(classifications_to_replace)
```

```python
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# review the dataFrame
application_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`

application_df = pd.get_dummies(application_df)
application_df.head()
```

```python
# winsorize the variable ASK_AMT

application_df["ASK_AMT_W"]= winsorize(application_df["ASK_AMT"], limits=(0.05, 0.95))
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df["IS_SUCCESSFUL"]
# drop target feature, unwinsorised column ASK_AMT, 
# and the column SPECIAL_CONSIDERATIONS_N (perfectly correlated to SPECIAL_CONSIDERATIONS_Y)
X = application_df.drop(columns=["IS_SUCCESSFUL", "ASK_AMT", "SPECIAL_CONSIDERATIONS_N"])
X.head()
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# check the shape of training data
X_train_scaled.shape
```

Optimise the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn = tf.keras.models.Sequential()

# Input layer
nn.add(tf.keras.layers.Dense(units=64, activation="relu", input_dim=42))

# First hidden layer
nn.add(tf.keras.layers.Dense(units=16, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=8, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
#Set initial batch size
initial_batch_size = 32

# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=80, batch_size = initial_batch_size)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
